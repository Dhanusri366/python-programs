```python
from operator import truediv

MENU = {
    "espresso": {
        "ingredients": {
            "water": 50,
            "coffee": 18,
            "milk": 0,
        },
        "cost": 1.5,
    },
    "latte": {
        "ingredients": {
            "water": 200,
            "milk": 150,
            "coffee": 24,
        },
        "cost": 2.5,
    },
    "cappuccino": {
        "ingredients": {
            "water": 250,
            "milk": 100,
            "coffee": 24,
        },
        "cost": 3.0,
    }
}

resources = {
    "water": 300,
    "milk": 200,
    "coffee": 100,
}
print(resources['water'])
money =0
```

```python
coffee = input("what coffee do you want (espresso/latte/cappuccino: ")
res = True
if(coffee== 'report'):
        print(f"Milk: {resources['milk']} \nWater: {resources['water']} \nCoffee: {resources['coffee']} \nMoney: {money}")
        coffee = input("what coffee do you want (espresso/latte/cappuccino: ")

    
if(resources["water"] > MENU[coffee]['ingredients']['water']):
    if(coffee != 'espresso'): 
        if (resources['milk']> MENU[coffee]['ingredients']['milk'] ):
            res = True 
        else:
                print("out of milk")
                res = False
else:
    print("out of water")
    res = False
if(resources['coffee']>MENU[coffee]['ingredients']['coffee']):
        res = True
else:
        print("out of coffee")
        res = False
while res == True:    
    print("enter coins")
    dimes = float(input("enter dimes:"))
    pennys = float(input("enter pennys:"))
    quaters = float(input("enter quaters:"))
    nickels= float(input("enter nickels:"))
    total = dimes*0.10 + quaters*0.25+nickels*0.05+pennys*0.01
    if(total >MENU[coffee]['cost']):
        money += MENU[coffee]['cost']
        resources['water']-= MENU[coffee]['ingredients']['water']
        resources['milk']-= MENU[coffee]['ingredients']['milk']
        resources['coffee']-= MENU[coffee]['ingredients']['coffee']
        print(f"your {coffee} is ready")
    
    
        balance = total-MENU[coffee]['cost']
        if(balance>0):
            print(f"your balance : {balance} ")
        break
    

    else:
        print("not enough coins, your money will be refunded.")
```
